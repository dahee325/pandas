## indexing, slicing, 조건 필터링
- 인덱스 접근을 하면 그 값은 데이터프레임이 아님

```python
df.loc[3] # location, 3번 인덱스에 있는 데이터 출력
```

```python
df.loc[3, 'class'] # 3번 인덱스에 있는 데이터 중 class 출력
```

```python
df.loc[3, ['class', 'age']] # 3번 인덱스에 있는 데이터 중 class, age 출력
```

```python
df.loc[2:5] # 2번째부터 5번째 데이터 출력, 마지막 데이터 포함
```

```python
df.loc[2:5, ['pclass', 'who']]
```

```python
df[['class', 'age']] # 컬럼을 하나만 출력할 때 결과의 타입이 시리즈지만, 대괄호를 하나 더 하면 데이터프레임으로 출력
```

```python
df.loc[:3, 'class':'deck'] # 컬럼으로도 슬라이싱 가능
```

```python
# 비교연산자 : True, False가 뭉쳐져있는 데이터를 만들어 주는 것
df['who'] == 'man' # who의 값이 'man'인 것을 출력하는 것이 아닌 'man'인지 아닌지 True, False로 나옴
```

```python
condition = df['who'] == 'man'
```

```python
# df[condition].head()
df[df['who'] == 'man'] # 같은 코드
```

```python
df.loc[condition].head() # 데이터프레임
```

```python
df.loc[condition]['age'] = 100
```

```python
df.loc[condition, 'age'] = 100 # condition 조건을 만족하는 데이터들 중에서 'age'의 값을 100으로 변환
```

```python
df.head()
```

```python
cond1 = df['fare'] > 30

cond2 = df['who'] == 'woman'
```

```python
cond2.head()
```

```python
df.loc[cond1 & cond2] # 조건 cond1와 cond2를 모두 만족하는 데이터 출력
```

```python
df.loc[cond1 | cond2].head(3) # 두개의 조건 중 하나만 맞아도 데이터 출력
```

```python
df.iloc[1, 3] # iloc는 []안에 무조건 숫자를 넣어야함
```

## Where()
- 조건이 False인 데이터들을 일괄적으로 수정하는 함수

```python
df.tail()
```

```python
cond = df['fare'] < 30
df['fare'].where(cond, 1) # 30이상인 데이터의 'fare'값을 1로 변환
```

### practice

```python
tips = sns.load_dataset('tips')
# tips.value_counts('day')
# tips.value_counts('time')
tips.value_counts('size')
```

- 남성 흡연자 조건 필터링
- tip을 가장 많이 낸 순서로 내림차순 정렬
- 상위 10개 출력

```python
cond1 = tips['sex'] == 'Male'
cond2 = tips['smoker'] == 'Yes'

tips.loc[cond1 & cond2].sort_values('tip', ascending=False).head(10)
```

- 식사인원이 5명 이상인 저녁테이블
- 흡연자가 없는 주말 테이블
- total_bill, tip 출력

```python
cond1 = tips['size'] >= 5
cond2 = tips['time'] == 'Dinner'
cond3 = tips['smoker'] == 'No'
cond4 = tips['day'].isin(['Sat', 'Sun'])

# tips.loc[(cond1 & cond2) & (cond3 & cond4)][['total_bill', 'tip']]
tips.loc[(cond1 & cond2) & (cond3 & cond4), ['total_bill', 'tip']]
# tips.loc[(cond1 & cond2) & (cond3 & cond4)].iloc[:, :2]
```
