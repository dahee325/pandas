```python
import seaborn as sns
import pandas as pd

df = sns.load_dataset('titanic')
```

```python
df.head(3)
```

```python
df_copy = df.copy()
```

## groupby()
하나의 컬럼을 기준으로 묶어서 데이터를 출력

```python
# 성별을 기준으로 데이터의 평균값
df_copy.groupby('sex').mean(numeric_only=True)
```

```python
df_copy.groupby(['sex', 'pclass']).mean(numeric_only=True) # 여러개의 컬럼을 기준으로 할 때는 [](리스트) 사용
```

```python
df_copy.groupby(['sex', 'pclass'])[['survived']].mean(numeric_only=True)
```

## reset_index()

```python
df_temp = df_copy.groupby(['sex', 'pclass'])[['survived']].mean()
```

```python
df_temp
```

```python
df_temp.reset_index()
```

```python
df_copy.groupby(['sex', 'pclass'])[['survived', 'age']].agg(['mean', 'sum']) # agg() : 여러 함수를 같이 보고싶을 때 사용
```

## pivot_table
- 각 컬럼이 어떠한 관계를 갖고 있는지, corr()와 비슷

```python
df_copy.pivot_table(index='who', values = 'survived')
```

```python
df_copy.pivot_table(index=['who', 'pclass'], values='survived')
```
