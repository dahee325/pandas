### 결측 데이터 필터링

```python
cond = df_copy['age'].isnull() # 결측치를 가진 사람들의 목록
df_copy.loc[cond, 'age'] = 30
df_copy.tail()
```

## fillna()

```python
df_copy = df.copy()
```

```python
df_copy.tail()
```

```python
# 결측치인 값을 100으로 바꿈
# 수정 결과를 저장하지 않고 결과만 출력
df_copy['age'] = df_copy['age'].fillna(100)
# 재할당해야 수정됨
```

```python
df_copy.tail()
```

```python
df_copy['deck'].value_counts()
```

```python
df_copy['deck'].fillna('C')
```

```python
df_copy['deck'].fillna('Z')
# 카테고리 데이터는 내가 지정하지 않은 데이터를 추가로 넣을 수 없음
```

```python
df_copy['deck'] = df_copy['deck'].cat.add_categories('Z') # 카테고리에 'Z'추가
df_copy['deck'].fillna('Z')
```

```python
age_mean = df_copy['age'].mean()
```

```python
df_copy['age'] = df_copy['age'].fillna(age_mean) # 결측치를 평균값으로 바꿈
```

```python
df_copy.info()
```

```python
df_copy = df.copy()
```

```python
df_copy['age'] = df_copy['age'].fillna(df_copy['age'].median())
```

```python
df_copy.tail()
```

### practice

- 남자 승객의 결측치는 남자 승객의 평균
- 여자 승객의 결측치는 여자 승객의 평균
- 결측치를 모두 채우고 나서 모든 사람의 평균

```python
df_copy = df.copy()
```

```python
cond1 = df_copy['sex'] == 'male'
male_mean = df_copy.loc[cond1, 'age'].mean()
print(male_mean)
```

```python
cond2 = df_copy['sex'] == 'female'
female_mean = df_copy.loc[cond2, 'age'].mean()
print(female_mean)
```

```python
df_copy.loc[cond1, 'age'] = df_copy.loc[cond1, 'age'].fillna(male_mean)
df_copy.loc[cond2, 'age'] = df_copy.loc[cond2, 'age'].fillna(female_mean)
```

```python
df_copy['age'].mean()
```
