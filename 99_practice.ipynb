```python
import pandas as pd
import seaborn as sns
```

## penguins
- species : 펭귄의 종류
- island : 펭귄이 사는 섬
- bill_length_mm : 부리의 길이
- bill_depth_mm : 부리의 깊이
- flipper_length_mm : 물갈퀴의 길이
- body_mass_g : 펭귄의 몸무게
- sex : 성별

```python
df = sns.load_dataset('penguins')
```

```python
df.head()
```

```python
# 5000g 이상의 펭귄 출력
cond = df['body_mass_g'] > 5000
df.loc[cond]
```

```python
# Gentoo종의 flipper_length_mm 평균
cond = df['species'] == 'Gentoo'
df.loc[cond, 'flipper_length_mm'].mean()
```

```python
# Female Chinstrap종의 bill_length_mm 높은순 정렬
cond1 = df['species'] == 'Chinstrap'
cond2 = df['sex'] == 'Female'
df.loc[cond1 & cond2].sort_values('bill_length_mm', ascending = False)
```

```python
# species(종)별 성별 구분하여 평균 몸무게 출력
# => ex) Adelie, Female / Adelie, male / Gentoo, Female...
```

```python
# 'Adelie'종인 펭귄의 평균 몸무게
cond1 = df['species'] == 'Adelie'
cond2 = df['sex'] == 'Male'
cond3 = df['sex'] == 'Female'

a_male = df.loc[cond1 & cond2, 'body_mass_g'].mean()
a_female = df.loc[cond1 & cond3, 'body_mass_g'].mean()
print(a_male)
print(a_female)
```

```python
#'Gentoo'종인 펭귄의 평균 몸무게
cond1 = df['species'] == 'Gentoo'
cond2 = df['sex'] == 'Male'
cond3 = df['sex'] == 'Female'

g_male = df.loc[cond1 & cond2, 'body_mass_g'].mean()
g_female = df.loc[cond1 & cond3, 'body_mass_g'].mean()
print(g_male)
print(g_female)
```

```python
# 'Chinstrap'종인 펭귄의 평균 몸무게
cond1 = df['species'] == 'Chinstrap'
cond2 = df['sex'] == 'Male'
cond3 = df['sex'] == 'Female'

c_male = df.loc[cond1 & cond2, 'body_mass_g'].mean()
c_female = df.loc[cond1 & cond3, 'body_mass_g'].mean()
print(c_male)
print(c_female)
```

```python
# bill_depth_mm 결측치 채우기 => 같은종 같은섬의 평균으로
cond = df['bill_depth_mm'].isnull() == True
df.loc[cond]

# 'Adelie'종 & 'Torgersen'섬
# 'Gentoo'종 & 'Biscoe'섬
```

```python
# bill_length_mm 결측치 채우기 => 같은종 같은섬의 평균으로
cond1 = df['species'] == 'Adelie'
cond2 = df['island'] == 'Torgersen'

a_bill_mean = df.loc[cond1 & cond2, 'bill_length_mm'].mean()
df.loc[cond1 & cond2, 'bill_length_mm'] = df.loc[cond1 & cond2, 'bill_length_mm'].fillna(a_bill_mean)

cond3 = df['species'] == 'Gentoo'
cond4 = df['island'] == 'Biscoe'

g_bill_mean = df.loc[cond3 & cond4, 'bill_length_mm'].mean()
df.loc[cond3 & cond4, 'bill_length_mm'] = df.loc[cond3 & cond4, 'bill_length_mm'].fillna(g_bill_mean)
```

```python
# flipper_length_mm 결측치 채우기 => 같은종 같은섬의 평균으로
cond1 = df['species'] == 'Adelie'
cond2 = df['island'] == 'Torgersen'

a_bill_mean = df.loc[cond1 & cond2, 'flipper_length_mm'].mean()
df.loc[cond1 & cond2, 'flipper_length_mm'] = df.loc[cond1 & cond2, 'flipper_length_mm'].fillna(a_bill_mean)

cond3 = df['species'] == 'Gentoo'
cond4 = df['island'] == 'Biscoe'

g_bill_mean = df.loc[cond3 & cond4, 'flipper_length_mm'].mean()
df.loc[cond3 & cond4, 'flipper_length_mm'] = df.loc[cond3 & cond4, 'flipper_length_mm'].fillna(g_bill_mean)
```

```python
# body_mass_g 결측치 채우기 => 같은종 같은섬의 평균으로
cond1 = df['species'] == 'Adelie'
cond2 = df['island'] == 'Torgersen'

a_bill_mean = df.loc[cond1 & cond2, 'body_mass_g'].mean()
df.loc[cond1 & cond2, 'body_mass_g'] = df.loc[cond1 & cond2, 'body_mass_g'].fillna(a_bill_mean)

cond3 = df['species'] == 'Gentoo'
cond4 = df['island'] == 'Biscoe'

g_bill_mean = df.loc[cond3 & cond4, 'body_mass_g'].mean()
df.loc[cond3 & cond4, 'body_mass_g'] = df.loc[cond3 & cond4, 'body_mass_g'].fillna(g_bill_mean)
```
