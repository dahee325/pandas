## merge()

```python
info = {
    '고객명': ['박세리', '이대호', '손흥민', '김연아', '마이클조던'],
    '생년월일': ['1980-01-02', '1982-02-22', '1993-06-12', '1988-10-16', '1970-03-03'],
    '성별': ['여자', '남자', '남자', '여자', '남자']
}

money = {
    '고객명': ['김연아', '박세리', '손흥민', '이대호', '타이거우즈'],
    '연봉': ['2000원', '3000원', '1500원', '2500원', '3500원']
}
```

```python
df1 = pd.DataFrame(info)
df2 = pd.DataFrame(money)
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2) # 두 데이터의 교집하바 데이터만 병합(inner merge)
```

```python
pd.merge(df1, df2, how='inner') # 교집합, 디폴트값
```

```python
pd.merge(df1, df2, how='outer') # 합집합, 비어있는 값은 NaN으로 채워줌
```

```python
pd.merge(df1, df2, how='left') # left에 있는 df1의 행에 df2 출력
```

```python
pd.merge(df1, df2, how='right') # right 있는 df2의 행에 df1 병합
```

```python
info = {
    '이름': ['박세리', '이대호', '손흥민', '김연아', '마이클조던'],
    '생년월일': ['1980-01-02', '1982-02-22', '1993-06-12', '1988-10-16', '1970-03-03'],
    '성별': ['여자', '남자', '남자', '여자', '남자']
}

money = {
    '고객명': ['김연아', '박세리', '손흥민', '이대호', '타이거우즈'],
    '연봉': ['2000원', '3000원', '1500원', '2500원', '3500원']
}
```

```python
df1 = pd.DataFrame(info)
df2 = pd.DataFrame(money)
```

```python
pd.merge(df1, df2, left_on='이름', right_on='고객명') # 왼쪽/오른쪽 각각 기준이 되는 컬럼의 이름 지정
```
