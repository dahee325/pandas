## 자전거 데이터 활용

```python
df = pd.read_csv('data/seoul_bicycle.csv')
df.head(2)
```

```python
df.info()
```

```python
df_copy = df.copy()
```

```python
df_copy['대여일자'] = pd.to_datetime(df_copy['대여일자'], format='%b-%d-%Y') # 내가 가지고 있는 정보의 객체를 datetime으로 변환
df_copy.info()
```

```python
cond = df_copy['대여일자'].dt.dayofweek == 3
df_copy.loc[cond]
```

```python
df_copy.head()
```

```python
pd.to_numeric(df_copy['운동량']) # 내가 가지고있는 컬럼을 숫자로 바꿈
```

```python
df_copy.loc[2344] # 이 하나의 데이터 때문에 오류
```

```python
df_copy['운동량'] = pd.to_numeric(df_copy['운동량'], errors='coerce') # 에러가 생겼을 때 처리하는 방법을 옵션으로 입력, 
# 'coerce' : 숫자로 바꿀 수 없는 값을 nan으로 지정
```

```python
df_copy['운동량']
```

### cut() 구간 나누기

```python
bins = [0, 100, 200, 300, df_copy['운동량'].max()]
pd.cut(df_copy['운동량'], bins)  #cut(나눌 대상의 데이터, 구간 데이터)
# 이상치가 나오므로 나중에 처리해야함
```

```python
# 자동으로 구간 나누기
pd.cut(df_copy['운동량'], bins=10)
pd.cut(df_copy['운동량'], bins=10).value_counts()
```

```python
pd.qcut(df_copy['운동량'], q=10) # qcut() : 10%(q=10)의 비율대로 나눔
pd.qcut(df_copy['운동량'], q=10).value_counts()
```

```python
qcut_bins = [0, 0.2, 0.8, 1] # 0~20%, 20~80%, 80~100%
df_copy['운동량_한글'] = pd.qcut(df_copy['운동량'], qcut_bins, labels=['적음', '보통', '많음']) # labels=[] : 나눈 구간의 이름을 설정
df_copy.head(2)
```
